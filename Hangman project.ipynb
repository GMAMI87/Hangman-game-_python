```python
#hangman in python
import random
#choose word list
word_list = ['avenue', 'beekeeper', 'cobweb', 'duplex', 'galaxy', 'jackport', 'luxury', 'nightclub']
print(word_list)
```

```python
# Choose a random word
chosen_word = random.choice(word_list)
print("Chosen word:", chosen_word)
```

```python
# Initialize display with dashes
display = ['_' for _ in chosen_word]
# Initialize variables
guessed_letters = []
failed_attempts = 0
max_attempts = 6
```

```python
# Game loop
while failed_attempts < max_attempts and '_' in display:
    # Show current word progress
    print("\nCurrent word:", ' '.join(display))
    
    # Ask user for a guess
    guess = input("Guess a letter: ").lower()
    
    # Check input validity
    if len(guess) != 1 or not guess.isalpha():
        print("Invalid input! Please enter a single letter.")
        continue
    
    # Check if letter was already guessed
    if guess in guessed_letters:
        print("You already guessed that letter. Try a different one.")
        continue
    
    # Add guess to guessed letters
    guessed_letters.append(guess)
    
    # Check if guess is in the chosen word
    if guess in chosen_word:
        print("Good guess!")
        # Reveal all positions of guessed letter
        for i, letter in enumerate(chosen_word):
            if letter == guess:
                display[i] = guess
    else:
        failed_attempts += 1
        print(f"Wrong guess! Attempts left: {max_attempts - failed_attempts}")
```

```python
# End of game messages
if '_' not in display:
   print("\nCongratulations! You guessed the word:", chosen_word)
else:  
   print("\nYou lost! The word was:", chosen_word)
   
```
