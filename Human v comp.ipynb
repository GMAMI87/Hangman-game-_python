```python
import random
def rock_paper_scissors_game(rounds=3):
    Options = ["rock", "paper", "scissors"]


    # Define the scores at the start of the game
    player_score = 0     # counts how many rounds the human player wins
    computer_score = 0   # counts how many rounds the computer wins


    print("Welcome to Rock-Paper-Scissors!")
    print(f"This game will be played for {rounds} rounds.")
    print("Rules: Rock beats Scissors, Scissors beats Paper, Paper beats Rock.\n")


    for round_num in range(1, rounds + 1):


        # Stop early if someone already won
        if (rounds == 3 and (player_score == 2 or computer_score == 2)) or \
        (rounds == 5 and (player_score == 3 or computer_score == 3)):
            break


        print(f"--- Round {round_num} ---")
        player_option = input("Enter your choice (rock, paper, scissors): ").lower()


        if player_option not in Options:
            print("Invalid choice! Please choose rock, paper, or scissors.")
            continue

        computer_options = random.choice(Options)
        print(f"Computer chose: {computer_options}")


    # Compare options and determine the winner
    if player_option == computer_options:
        print("It's a tie!")
    elif (player_option == "rock" and computer_options == "scissors") or \
        (player_option == "scissors" and computer_options == "paper") or \
        (player_option == "paper" and computer_options == "rock"):
       print("You win this round!")
       player_score += 1
    else:
        print("Computer wins this round!")
        computer_score += 1


    print(f"Score -> You: {player_score}, Computer: {computer_score}\n")


 # RETURN the scores 
    return player_score, computer_score



#Final result
player_score, computer_score = rock_paper_scissors_game(3)

print("\n=== Final Result ===")
if player_score > computer_score:
    print("Congratulations! You won the game!")
elif computer_score > player_score:
    print("Computer wins the game! Better luck next time.")
else:
    print("The game is a tie!")




        

```
